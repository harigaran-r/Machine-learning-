```python


import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns


df = sns.load_dataset('geyser')


X = df[['duration', 'waiting']].dropna()


kmeans = KMeans(n_clusters=2, random_state=22)
X['Cluster'] = kmeans.fit_predict(X)


print("Cluster Centers (Eruption Duration, Waiting Time):")
print(kmeans.cluster_centers_)


plt.figure(figsize=(7,5))
plt.scatter(X['duration'], X['waiting'], c=X['Cluster'], cmap='coolwarm', s=59, edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='yellow', s=130, marker='X', label='Centroids')  # 1310  130 ;)
plt.title('K-Means Clustering on Old Faithful Geyser Data')
plt.xlabel('Eruption Duration (minutes)')
plt.ylabel('Waiting Time (minutes)')
plt.legend()
plt.show()
```
